FIR Hamming window low pass

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter lowpass FIR
def design_lowpass_fir(numtaps, cutoff, fs):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    taps = signal.firwin(numtaps, norm_cutoff, window='hamming', pass_zero=True)
    return taps

# Baca file WAV
input_wav_path = 'Av7 - Afterlife1.wav'
output_wav_path = 'output_FIR_hamming_lowpass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
numtaps = 101  # Panjang filter
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
taps = design_lowpass_fir(numtaps, cutoff_freq, fs)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(taps, 1.0, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
plt.subplot(4, 1, 3)
plt.plot(taps)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(taps, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5 * fs * w / np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

Filter IIR Bessel

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter lowpass IIR dengan jendela Bessel
def design_lowpass_iir(cutoff, fs, order):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    b, a = signal.bessel(order, norm_cutoff, btype='low', analog=False)
    return b, a

# Baca file WAV
input_wav_path = 'Av7 - Afterlife1.wav'
output_wav_path = 'output_IIR_bessel_lowpass.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter
cutoff_freq = 1000  # Frekuensi cutoff dalam Hz
order = 4  # Orde filter
b, a = design_lowpass_iir(cutoff_freq, fs, order)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
impulse = np.zeros(100)
impulse[0] = 1
impulse_response = signal.lfilter(b, a, impulse)
plt.subplot(4, 1, 3)
plt.plot(impulse_response)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5 * fs * w / np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt

# Fungsi untuk membuat filter notch IIR
def design_notch_iir(cutoff, fs, Q):
    nyquist = 0.5 * fs
    norm_cutoff = cutoff / nyquist
    b, a = signal.iirnotch(norm_cutoff, Q, fs)
    return b, a

# Baca file WAV
input_wav_path = 'Av7 - Afterlife1.wav'
output_wav_path = 'output_IIR_notch.wav'
fs, data = wavfile.read(input_wav_path)

# Pastikan data adalah array 1D
if data.ndim > 1:
    data = data[:, 0]

# Desain filter notch
notch_freq = 1000  # Frekuensi notch dalam Hz
Q = 30  # Faktor kualitas (Quality factor)
b, a = design_notch_iir(notch_freq, fs, Q)

# Terapkan filter ke sinyal
filtered_data = signal.lfilter(b, a, data)

# Simpan hasil sinyal terfilter ke file WAV
wavfile.write(output_wav_path, fs, filtered_data.astype(np.int16))

# Plot sinyal asli dan hasil filter
plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(data)
plt.title('Sinyal Asli')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(4, 1, 2)
plt.plot(filtered_data)
plt.title('Sinyal Setelah Filter Notch')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# Plot impulse response
impulse = np.zeros(100)
impulse[0] = 1
impulse_response = signal.lfilter(b, a, impulse)
plt.subplot(4, 1, 3)
plt.plot(impulse_response)
plt.title('Impulse Response')
plt.xlabel('Tap')
plt.ylabel('Amplitude')

# Plot frequency response
w, h = signal.freqz(b, a, worN=8000)
plt.subplot(4, 1, 4)
plt.plot(0.5 * fs * w / np.pi, np.abs(h))
plt.title('Frequency Response')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain')

plt.tight_layout()
plt.show()
```
